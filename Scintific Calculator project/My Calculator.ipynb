```python
from tkinter import *
from tkinter.messagebox import *
import math as m
font = ('verdana',20,'bold')

#important function
def clear():
    ex1 = textField.get()
    ex1 = ex1[0:len(ex1)-1]
    textField.delete(0,END)
    textField.insert(0,ex1)

def all_clear():
    textField.delete(0,END)

def click_button_fun(event):
    b = event.widget
    text = b['text']
    
    if text=='x':
        textField.insert(END , '*')
        return 
    if text=='=':
        try:
            ex = textField.get()
            answer = eval(ex)
            textField.delete(0 , END)
            textField.insert(0 , answer)
        except Exception as e:
            print('error ..',e)
            showerror('Erroe..',e)
        return
    
    textField.insert(END,text)
    
window = Tk()
window.title('my calculator')
window.geometry('500x500')
#image label
#pic = PhotoImage(pathof image)
headingLabel = Label(window,text="my calcy",font = font, underline=0)
headingLabel.pack(side=TOP)

#textfield
textField = Entry(window , font=font,justify=CENTER)
textField.pack(side=TOP , pady=10 ,padx=10, fill=X)

#buttens
buttonFrame = Frame(window)
buttonFrame.pack(side=TOP)

#addingbutton
temp = 1
for i in range(0,3):
    for j in range(0,3):
        btn=Button(buttonFrame , text=str(temp),font=font,width=5,relief='ridge',activebackground='orange',activeforeground='white') 
        btn.grid(row=i , column=j,padx=5,pady=5)
        temp=temp+1
        btn.bind('<Button-1>',click_button_fun)
        
btn1=Button(buttonFrame,text='0',font=font,width=5,relief='ridge',activebackground='orange',activeforeground='white')
btn1.grid(row=3 , column=0,padx=5,pady=5)

btn2=Button(buttonFrame,text='.',font=font,width=5,relief='ridge',activebackground='orange',activeforeground='white')
btn2.grid(row=3 , column=1,padx=5,pady=5)

btn3=Button(buttonFrame , text='=',font=font,width=5,relief='ridge',activebackground='orange',activeforeground='white')
btn3.grid(row=3 , column=2,padx=5,pady=5)

btn4=Button(buttonFrame , text='+',font=font,width=5,relief='ridge',activebackground='orange',activeforeground='white')
btn4.grid(row=0 , column=3,padx=5,pady=5)

btn5=Button(buttonFrame , text='-',font=font,width=5,relief='ridge',activebackground='orange',activeforeground='white')
btn5.grid(row=1 , column=3,padx=5,pady=5)

btn6=Button(buttonFrame , text='x',font=font,width=5,relief='ridge',activebackground='orange',activeforeground='white')
btn6.grid(row=2 , column=3,padx=5,pady=5)

btn7=Button(buttonFrame , text='/',font=font,width=5,relief='ridge',activebackground='orange',activeforeground='white')
btn7.grid(row=3 , column=3,padx=5,pady=5)

btn8=Button(buttonFrame , text='<--',font=font,width=5,relief='ridge',activebackground='orange',activeforeground='white',command=clear)
btn8.grid(row=4 , column=0,padx=5,pady=5,columnspan=2)

btn9=Button(buttonFrame , text='AC',font=font,width=5,relief='ridge',activebackground='orange',activeforeground='white',command=all_clear)
btn9.grid(row=4 , column=2,padx=5,pady=5,columnspan=2)

#binding all buttons
btn1.bind('<Button-1>',click_button_fun)
btn2.bind('<Button-1>',click_button_fun)
btn3.bind('<Button-1>',click_button_fun)
btn4.bind('<Button-1>',click_button_fun)
btn5.bind('<Button-1>',click_button_fun)
btn6.bind('<Button-1>',click_button_fun)
btn7.bind('<Button-1>',click_button_fun)
#btn8.bind('<Button-1>',all_clear)
#btn9.bind('<Button-1>',all_clear)

def enterClick(event):
    e = Event()
    e.widget = btn3
    click_button_fun(e)
window.bind('<Return>',enterClick)

#menubar #########################################################################################
scFrame = Frame(window)

btn10=Button(scFrame , text='sqrt',font=font,width=5,relief='ridge',activebackground='orange',activeforeground='white')
btn10.grid(row=0 , column=0,padx=5, pady=5)

btn11=Button(scFrame , text='^',font=font,width=5,relief='ridge',activebackground='orange',activeforeground='white')
btn11.grid(row=0 , column=1,padx=5, pady=5)

btn12=Button(scFrame , text='x!',font=font,width=5,relief='ridge',activebackground='orange',activeforeground='white')
btn12.grid(row=0 , column=2,padx=5, pady=5)

btn13=Button(scFrame , text='toRad',font=font,width=5,relief='ridge',activebackground='orange',activeforeground='white')
btn13.grid(row=0 , column=3,padx=5, pady=5)

btn14=Button(scFrame , text='toDec',font=font,width=5,relief='ridge',activebackground='orange',activeforeground='white')
btn14.grid(row=1 , column=0,padx=5, pady=5)

btn15=Button(scFrame , text='sin0',font=font,width=5,relief='ridge',activebackground='orange',activeforeground='white')
btn15.grid(row=1 , column=1,padx=5, pady=5)

btn16=Button(scFrame , text='cos0',font=font,width=5,relief='ridge',activebackground='orange',activeforeground='white')
btn16.grid(row=1 , column=2,padx=5, pady=5)

btn17=Button(scFrame , text='tan0',font=font,width=5,relief='ridge',activebackground='orange',activeforeground='white')
btn17.grid(row=1 , column=3,padx=5, pady=5)

normalcalc = True

def calculator_sc(event):
    print('clicked')
    btn=event.widget
    text=btn['text']
    print('text')
    #textField.insert(0 , text)
    ex = textField.get()
    answer=''
    if text=='sqrt':
        print('sqrt is')
        answer = str(m.sqrt(float(ex)))   
        
    elif text=='^':
        print('power is')
        base,power = ex.split(',')
        answer = str(m.pow(float(base),float(power)))

    elif text=='x!':
        print('fact is')
        answer = str(m.factorial(int(ex)))

    elif text=='toRad':
        print('Radian is')
        answer = str(m.radians(float(ex)))

    elif text=='toDec':
        print('Decimal is')
        answer = str(m.degrees(float(ex)))
        
    elif text=='sin0':
        print('sin0 is')
        answer = str(m.sin(m.radians(int(ex))))

    elif text=='cos0':
        print('cos0 is')
        answer = str(m.cos(m.radians(int(ex))))

    elif text=='tan0':
        print('tan0 is')
        answer = str(m.tan(m.radians(int(ex))))

    textField.delete(0,END)
    textField.insert(0,answer)

def sc_click():
    global normalcalc
    if normalcalc:
        buttonFrame.pack_forget()
        scFrame.pack(side = TOP , pady=10)
        
        buttonFrame.pack(side = TOP)
        window.geometry('500x600')
        print('show scintific')
        normalcalc = False
    else:
        print('show normal')
        scFrame.pack_forget()
        window.geometry('500x500')
        normalcalc = True

#bind scbuttons
btn10.bind('<Button-1>',calculator_sc)
btn11.bind('<Button-1>',calculator_sc)
btn12.bind('<Button-1>',calculator_sc)
btn13.bind('<Button-1>',calculator_sc)
btn14.bind('<Button-1>',calculator_sc)
btn15.bind('<Button-1>',calculator_sc)
btn16.bind('<Button-1>',calculator_sc)
btn17.bind('<Button-1>',calculator_sc)

        
font2=('',15)
menubar = Menu(window)
mode = Menu(menubar ,font=font2,tearoff=0)
mode.add_checkbutton(label="Scientific Calculator",command=sc_click)
menubar.add_cascade(label='mode',menu=mode)

window.config(menu=menubar)

window.mainloop()
```
